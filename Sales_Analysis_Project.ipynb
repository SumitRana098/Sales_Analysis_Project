**Reasoning**:
Import the pandas library and read the CSV file into a DataFrame.



```python
import pandas as pd
import os
from google.colab import files

# List files in the current directory to verify the file upload
print(os.listdir('.'))

# Replace 'your_uploaded_file_name.csv' with the actual name of the uploaded file
# and read the CSV file into a pandas DataFrame
try:
    df = pd.read_csv('Sales Data.csv')
    # Display the first few rows of the DataFrame
    display(df.head())
except FileNotFoundError:
    print("Error: 'Sales Data.csv' not found. Please check the uploaded filename.")
```

```python
# Display the first few rows of the DataFrame
display(df.head())

# Get information about the DataFrame, including data types and non-null values
display(df.info())

# Drop the 'Unnamed: 0' column
df = df.drop('Unnamed: 0', axis=1)

# Display the first few rows and info again to confirm changes
display(df.head())
display(df.info())
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
display(df.info())
```

```python
# Check for missing values
print("Missing values before cleaning:")
display(df.isnull().sum())

# Drop rows with missing values
df.dropna(inplace=True)

# Check for duplicate rows
print("\nDuplicate rows before cleaning:")
display(df.duplicated().sum())

# Drop duplicate rows
df.drop_duplicates(inplace=True)

# Verify changes
print("\nMissing values after cleaning:")
display(df.isnull().sum())

print("\nDuplicate rows after cleaning:")
display(df.duplicated().sum())
```

```python
# Extract Month and Day from 'Order Date'
df['Day'] = df['Order Date'].dt.day

# Calculate 'Total Sales'
df['Total Sales'] = df['Quantity Ordered'] * df['Price Each']

# Extract City from 'Purchase Address'
# Assuming the city is the second element when splitting by comma
df['City'] = df['Purchase Address'].apply(lambda x: x.split(',')[1].strip())

# Display the first few rows to show the new columns
display(df.head())
```

```python
import calendar

df['Month Name'] = df['Month'].apply(lambda x: calendar.month_name[x])
display(df.head())
```

```python
# Summarize key statistics
print("Key Statistics:")
display(df[['Quantity Ordered', 'Price Each', 'Sales', 'Total Sales']].describe())

# Analyze Sales by Month
print("\nTotal Sales by Month:")
monthly_sales = df.groupby('Month Name')['Total Sales'].sum().sort_index()
display(monthly_sales)

# Analyze Sales by City
print("\nTotal Sales by City:")
city_sales = df.groupby('City')['Total Sales'].sum().sort_values(ascending=False)
display(city_sales)

# Analyze Most Popular Products (by Quantity Ordered)
print("\nTop 10 Most Popular Products:")
product_sales = df.groupby('Product')['Quantity Ordered'].sum().sort_values(ascending=False).head(10)
display(product_sales)

# Analyze Sales by Hour
print("\nTotal Sales by Hour:")
hourly_sales = df.groupby('Hour')['Total Sales'].sum().sort_index()
display(hourly_sales)
```

## Data Visualization

Let's create some visualizations to better understand the sales trends.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import matplotlib.colors as mcolors

# Set the color theme for seaborn
sns.set_palette("Blues")

# Get a color from the Seaborn palette and convert it to a hex string for Plotly
seaborn_color = sns.color_palette("Blues")[0]
plotly_color = mcolors.to_hex(seaborn_color)


# Visualize Sales by Month using Matplotlib/Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x=monthly_sales.index, y=monthly_sales.values)
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualize Sales by City using Plotly
# Using a Plotly template that might be visually similar to Blues
fig = px.bar(city_sales, x=city_sales.index, y=city_sales.values, labels={'x': 'City', 'y': 'Total Sales'}, title='Total Sales by City', template='plotly_white')
fig.update_layout(colorway=[plotly_color] * len(city_sales.index)) # Applying the same color from the Blues palette
fig.show()

# Visualize Top 10 Most Popular Products using Matplotlib/Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x=product_sales.index, y=product_sales.values)
plt.title('Top 10 Most Popular Products (by Quantity Ordered)')
plt.xlabel('Product')
plt.ylabel('Quantity Ordered')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualize Sales by Hour using Matplotlib/Seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(x=hourly_sales.index, y=hourly_sales.values)
plt.title('Total Sales by Hour')
plt.xlabel('Hour')
plt.ylabel('Total Sales')
plt.xticks(hourly_sales.index)
plt.grid(True)
plt.show()
```

## Investigate price and quantity relationship

### Subtask:
Examine the relationship between 'Price Each' and 'Quantity Ordered', perhaps through scatter plots or correlation analysis.


**Reasoning**:
Calculate and display the correlation between 'Price Each' and 'Quantity Ordered', then generate a scatter plot to visualize this relationship.



```python
# Calculate the correlation coefficient
correlation = df['Price Each'].corr(df['Quantity Ordered'])
print(f"Correlation between Price Each and Quantity Ordered: {correlation}")

# Generate a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Price Each', y='Quantity Ordered')
plt.title('Relationship between Price Each and Quantity Ordered')
plt.xlabel('Price Each')
plt.ylabel('Quantity Ordered')
plt.show()
```

## Analyze sales trends over time

### Subtask:
Look for seasonality and trends by analyzing sales over time, potentially by grouping by week, month, or quarter.


**Reasoning**:
Group the data by order date and resample to monthly, quarterly, and weekly frequencies to analyze sales trends over time, then display the head of each resulting Series.



```python
monthly_trend = df.groupby('Order Date')['Total Sales'].sum().resample('M').sum()
quarterly_trend = df.groupby('Order Date')['Total Sales'].sum().resample('Q').sum()
weekly_trend = df.groupby('Order Date')['Total Sales'].sum().resample('W').sum()

print("Monthly Sales Trend:")
display(monthly_trend.head())

print("\nQuarterly Sales Trend:")
display(quarterly_trend.head())

print("\nWeekly Sales Trend:")
display(weekly_trend.head())
```
