## Analyze trends in netflix content over time

### Subtask:
Visualize the trend of content added over time, differentiating between movies and TV shows.


**Reasoning**:

```python
movies_df = df1[df1['type'] == 'Movie'].copy()
tv_shows_df = df1[df1['type'] == 'TV Show'].copy()

movies_added_over_time = movies_df.groupby(['year_added', 'month_added']).size().reset_index(name='count')
movies_added_over_time['type'] = 'Movie'

tv_shows_added_over_time = tv_shows_df.groupby(['year_added', 'month_added']).size().reset_index(name='count')
tv_shows_added_over_time['type'] = 'TV Show'

content_added_over_time = pd.concat([movies_added_over_time, tv_shows_added_over_time])

content_added_over_time['date'] = pd.to_datetime(content_added_over_time[['year_added', 'month_added']].assign(day=1))

content_added_over_time = content_added_over_time.sort_values(by=['year_added', 'month_added'])

plt.figure(figsize=(14, 7))
sns.lineplot(data=content_added_over_time, x='date', y='count', hue='type')
plt.title('Netflix Content Added Over Time by Type')
plt.xlabel('Date Added')
plt.ylabel('Number of Titles Added')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/Netflix.csv')

# Create a copy of the DataFrame
df1 = df.copy()

# --- Data Cleaning ---

# Handle missing values by filling with 'Unknown'
df1['director'].fillna('Unknown', inplace=True)
df1['cast'].fillna('Unknown', inplace=True)
df1['country'].fillna('Unknown', inplace=True)

# Convert 'date_added' to datetime objects
df1['date_added'] = pd.to_datetime(df1['date_added'], format='%d-%b-%y')

# Split into Movies and TV Shows DataFrames
movies_df = df1[df1['type'] == 'Movie'].copy()
tv_shows_df = df1[df1['type'] == 'TV Show'].copy()

# Clean and convert 'duration' for movies
movies_df['duration'] = movies_df['duration'].str.replace(' min', '').astype(int)
movies_df.rename(columns={'duration': 'duration_min'}, inplace=True)

# Remove duplicates (checked and found none in previous steps, but good practice)
df1.drop_duplicates(inplace=True)

# --- Data Analysis ---

# Analyze trends in content added over time
df1['month_added'] = df1['date_added'].dt.month
df1['year_added'] = df1['date_added'].dt.year

movies_added_over_time = df1[df1['type'] == 'Movie'].groupby(['year_added', 'month_added']).size().reset_index(name='count')
movies_added_over_time['type'] = 'Movie'

tv_shows_added_over_time = df1[df1['type'] == 'TV Show'].groupby(['year_added', 'month_added']).size().reset_index(name='count')
tv_shows_added_over_time['type'] = 'TV Show'

content_added_over_time = pd.concat([movies_added_over_time, tv_shows_added_over_time])
content_added_over_time['date'] = pd.to_datetime(content_added_over_time[['year_added', 'month_added']].rename(columns={'year_added': 'year', 'month_added': 'month'}).assign(day=1))
content_added_over_time = content_added_over_time.sort_values(by=['year_added', 'month_added'])

# Visualize content added over time
plt.figure(figsize=(14, 7))
sns.lineplot(data=content_added_over_time, x='date', y='count', hue='type')
plt.title('Netflix Content Added Over Time by Type')
plt.xlabel('Date Added')
plt.ylabel('Number of Titles Added')
plt.show()

# Understand genre popularity
genres = df1['listed_in'].str.split(', ').explode()
top_genres = genres.value_counts().head(10)

# Visualize genre distribution
plt.figure(figsize=(12, 6))
top_genres.plot(kind='bar', color=sns.color_palette('viridis', 10))
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Examine the distribution of content across different regions
countries = df1['country'].str.split(', ').explode()
top_countries = countries.value_counts().head(10)

# Visualize content distribution by country
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='viridis')
plt.title('Top 10 Countries by Content on Netflix')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Examine the distribution of content across different time periods (release year)
release_year_counts = df1['release_year'].value_counts().sort_index()

# Visualize release year distribution
plt.figure(figsize=(12, 6))
release_year_counts.plot(kind='line', marker='o')
plt.title('Distribution of Content Release Years on Netflix')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.show()

# Combine insights (Example: Genre distribution over release years, Country vs Type, Rating vs Release Year)
genres_over_time = df1.assign(listed_in=df1['listed_in'].str.split(', ')).explode('listed_in')
genre_release_year_distribution = genres_over_time.groupby(['release_year', 'listed_in']).size().unstack(fill_value=0)
display("Distribution of Genres over Release Years (last 5 years):")
display(genre_release_year_distribution.tail())

country_type_distribution = df1[df1['country'].isin(top_countries.index)].groupby(['country', 'type']).size().unstack(fill_value=0)
display("\nRelationship between Top Countries and Content Type:")
display(country_type_distribution)

rating_release_year_distribution = df1.groupby(['release_year', 'rating']).size().unstack(fill_value=0)
display("\nDistribution of Ratings over Release Years (last 5 years):")
display(rating_release_year_distribution.tail())

# Example of exploring relationships: Average movie duration over the years
# Need to re-create movies_df and clean duration as it was done in a separate cell previously
movies_df_cleaned = df1[df1['type'] == 'Movie'].copy()
movies_df_cleaned['duration'] = movies_df_cleaned['duration'].str.replace(' min', '').astype(int)
movies_df_cleaned.rename(columns={'duration': 'duration_min'}, inplace=True)

average_movie_duration_over_years = movies_df_cleaned.groupby('release_year')['duration_min'].mean()
display("\nAverage Movie Duration over Release Years:")
display(average_movie_duration_over_years.tail())

plt.figure(figsize=(12, 6))
average_movie_duration_over_years.plot(kind='line', marker='o')
plt.title('Average Movie Duration Over Release Years')
plt.xlabel('Release Year')
plt.ylabel('Average Duration (minutes)')
plt.grid(True)
plt.show()
```
